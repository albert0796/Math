# 1-1 將 Example 2.37 之運算的疊代結果儲存起來

```python
def gaussian_seidel(x2,n):
    list_x1=[]
    list_x2=[]
    for i in range(n):
        x1=x2
        x2=-0.5*x1+1
        list_x1.append(x1)
        list_x2.append(x2)
    return list_x1,list_x2
x1,x2=gaussian_seidel(0,10) 
print(x1,x2)
```

# HW 1-2將 Example 2.37 疊代的收斂結果繪製出來

```python
%matplotlib inline 
import matplotlib.pyplot as plt
def gaussian_seidel(x2,n):
    list_x1=[]
    list_x2=[]
    for i in range(n):
        x1=x2
        x2=-0.5*x1+1
        list_x1.append(x1)
        list_x2.append(x2)
    return list_x1,list_x2
x1,x2=gaussian_seidel(0,10)        
plt.plot(range(10),x1,label='x1')
plt.plot(range(10),x2,label='x2')
plt.legend(loc='best')
```

# HW2：代入 Example 2.38 (p.127) 不同的 Initial Guess 看是否能使線性方程組的解由發散變成收斂

```python
import matplotlib.pyplot as plt

def gaussian_seidel(x2,n):
    list_x1=[]
    list_x2=[]
    for i in range(n):
        x1=1+x2
        x2=5-2*x1
        list_x1.append(x1)
        list_x2.append(x2)
    return list_x1,list_x2
x1,x2=gaussian_seidel(3,10)

plt.figure(figsize=(10,5))
plt.plot(range(10),x1,label='x1')
plt.plot(range(10),x2,label='x2')
plt.legend(loc='best')
plt.show()
```

# HW 3 實作矩陣乘法

```python
import numpy as np
a=np.mat([[1,3,-1],
        [-2,-1,1]])
b=np.mat([[-4,0,3,-1],
          [5,-2,-1,1],
          [-1,2,8,6]])
dot=a*b

c=np.zeros((a.shape[0],b.shape[1])).astype(np.int32)
for i in range(a.shape[0]):
    for j in range(b.shape[1]):
        temp=0
        for k in range(a.shape[1]):
            temp=temp+a[i,k]*b[k,j]
            c[i,j]=temp
print(dot)
print(c)
```
